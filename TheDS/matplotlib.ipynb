```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=list(np.random.randint(1,10,(10,1)))
y=list(np.random.randint(1,10,(10,1)))

plt.figure(1)#optional
plt.scatter(x,y)#it add one by one components to the fig
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("1st fig")
plt.show()#it shows the fig and deactivate the fig for farther editing

plt.figure(2)#optional
plt.plot(x,y)
plt.title("2nd fig")
plt.show()
```

```python
#BAR plot
x=["a","b","c","d"]
y=[4,3,6,1]

plt.bar(x,y)
```
